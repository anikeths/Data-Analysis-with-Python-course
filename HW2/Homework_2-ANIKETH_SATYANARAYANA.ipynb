```python
import pandas as pd
import numpy as np
```

```python
tables = pd.read_html('https://en.wikipedia.org/wiki/College_admissions_in_the_United_States')
```

    From the data in the above link find:
    1. weighted average of admission rate
    2. Average admission rate by State
    3. How many schools get between (1-10K), (10K-20K), etc. applications


## weighted average of admission rate

```python
# To select the 0-indexed table from the Wikipedia/URL 
df1= pd.DataFrame(tables[0])

# To select the column of interest
df2= df1['Total (56 institutions)']

# Refining column names
df2.columns = [''] * len(df2.columns)
df2.columns= ['No_Of_Applications','Admits','Enrollments','Admit Rate','Admits:Enrollments']

# To replacing/change the Admit Rate column format
df2['Admit Rate'].replace('%','', inplace= True, regex=True)
df2['Admit Rate']= df2['Admit Rate'].astype('float64')

display(df2)

# To calculate the weighted average
data_to_weight = df2['Admit Rate']
weights = df2['No_Of_Applications']

def weighted_average(data_to_weight, weights):
    num = sum(df2['Admit Rate']*df2['No_Of_Applications'])
    den = sum(df2['No_Of_Applications'])
    
    return round(num/den,2)

print('The Weighted Average is : {}'.format(weighted_average(data_to_weight, weights)))
```

## How many schools get between (1-10K), (10K-20K), etc. applications

```python
# To select the table of interest
schools = pd.DataFrame(tables[4])
schools=schools.drop(['No.Of Applications in this Range'],axis=1)
schools.head()
```

```python
# Binning
schools['Binned_ranges'] = pd.cut(x = schools.Applicants, 
                bins = [0, 10000, 20000, 30000, 40000, 50000, 60000, 70000, 80000, 90000, 100000, 110000,120000], 
                        )
schools.head()
```

```python
# To find the Number of schools in each binned-range

schools =schools.groupby('Binned_ranges')['School'].count()
print(schools)
```
